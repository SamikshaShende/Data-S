````python
# prompt: how to remove the duplicate and make a new list

def remove_duplicates(input_list):
    """Removes duplicates from a list while preserving order.

    Args:
        input_list: The list to remove duplicates from.

    Returns:
        A new list with duplicates removed, maintaining the original order.
    """
    unique_list = []
    seen = set()  # Use a set to efficiently check for duplicates

    for item in input_list:
        if item not in seen:
            unique_list.append(item)
            seen.add(item)
    return unique_list


# Example usage with the given list 'lst':
lst = [1, 2, 4, 2, 5, 1, 7, 12, 3, 78, 8, 4, 5, 6]
new_list = remove_duplicates(lst)
print(f"Original list: {lst}")
print(f"List with duplicates removed: {new_list}")


# Example Usage with a mixed list
my_list = [1, "apple", 3.14, True, "apple", 1, 1.0] #Note: 1 and 1.0 are considered different
new_list2 = remove_duplicates(my_list)
print(f"Original list: {my_list}")
print(f"List with duplicates removed: {new_list2}")
````

````python
#rewrite factorial as a function

def factorial(n):
  """Duck string : This Function calculates the factorial of a given argument"""

  if n<0:
    return None #"Factorial is not defined for negative numbers"
  elif (type(n)== float):
    return "Factorial is not defined for float numbers"
  else:
    f =1
    for i in range(1,n+1):
      f *=i

  return f
````

````python
#Fibonacci in a function
def fib_seq(n):

  fibonacci_Series = [0,1]
  if n>=2:
    for i in range (2,n+1):
      fseq= fibonacci_Series[i-1]+fibonacci_Series[i-2]
      fibonacci_Series.append(fseq)

  elif n == 0 or n == 1:
      fibonacci_Series = fseq[:n+1]
  #elif n<0:
    #return None
  else:
      fibonacci_Series = [None]
  return fibonacci_Series
````

````python
def fibo(n):
  if n==0 or n==1:
    return n
  elif n<0:
    return None
  else:
    return fibo(n-1)+fibo(n-2)


````
